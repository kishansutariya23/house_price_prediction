```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# import dataset
# download dataset from :->https://www.kaggle.com/datasets/ashydv/housing-dataset
data=pd.read_csv("./Housing.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
# select price and area
area=data['area']
price=data['price']
```

```python
area
```

```python
price
```

```python
#convert area into 2d matrix
X=np.array(area).reshape(-1,1)
X
```

```python
#convert price into 2d matrix
y=np.array(price).reshape(-1,1)
y
```

```python
# convert the data into test and training
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,y,test_size=1/3) 
```

```python
 # pass data into linear regression model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
# create predictor
predictor = regressor.predict(x_test)
predictor
```

```python
# this prediction is more helpful with graph
# plot for training dataset
plt.scatter(x_train,y_train)
plt.plot(x_train,regressor.predict(x_train),color='red')
plt.title("Training graph for Housing")
```

```python
# plot for test dataset
plt.scatter(x_test,y_test)
plt.plot(x_train,regressor.predict(x_train),color='red')
plt.title("Training graph for Housing")
```
